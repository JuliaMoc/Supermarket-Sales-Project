## Preprocessing

```python
# Changing the data type from object to datetime for the 'Date' column
df.Date = pd.to_datetime(df.Date)
```

```python
# Check if it worked
df.info()
```

```python
# Now I can easily extract datetime information from the 'Date' column
df.Date.dt.year
```

```python
df.Date.dt.month
```

```python
df.Date.dt.weekday
```

```python
# Renaming almost all column names
df = df.rename(columns = {'Invoice ID':'invoice_id',
                            'Branch':'branch',
                            'City':'city',
                            'Customer type':'customer_type',
                            'Gender':'gender',
                            'Product line':'product_category',
                            'Unit price':'unit_price',
                            'Quantity':'quantity',
                            'Tax 5%':'tax_5%',
                            'Total':'total',
                            'Date':'purchase_date',
                            'Time':'time',
                            'Payment':'type_of_payment',
                            'gross margin percentage':'gross_margin_percentage',
                            'gross income':'gross_income',
                            'Rating':'rating'
                           })
```

```python
# Checking if it worked
df
```

```python
# Drop column: branch, tax_5%, cogs, gross_margin_percentage
df = df.drop(['branch', 'tax_5%', 'cogs', 'gross_margin_percentage', 'gross_income'], axis = 1)
```

```python
df
```

## Data Analysis

```python
# Total Sales by Supermarket:
df.total.sum()
```

```python
# Total Sales by Supermarket:
df.groupby('city').agg('total').sum()
```

```python
display(df.groupby('city').agg('total').sum().sort_values(ascending=False));

df.groupby('city').agg('total').sum().sort_values(ascending=False).plot(kind='bar', color='#66D1C5');
plt.title('Total Sales by Supermarket');
```

```python
# Total and Avg. Sales by Product Category:
df.groupby('product_category').agg({'total':['sum', 'mean']})
```

```python
display(df.groupby('product_category').agg('total').sum().sort_values(ascending=False));

df.groupby('product_category').agg('total').sum().sort_values(ascending=False).plot(kind='bar', color='#66D1C5');
plt.title('Total Sales by Product Category');
```

```python
# Total Sales per Day:
df.groupby('purchase_date').agg('total').sum()
```

```python
display(df.groupby('purchase_date').agg('total').sum().sort_values(ascending=False));

df.groupby('purchase_date').agg('total').sum().sort_values(ascending=False).plot(kind='line', color='#66D1C5');
plt.title('Total Sales per Day');
```

```python
# Total Quantity:
df.quantity.sum()
```

```python
# Avg. Rating per Day:
df.groupby('purchase_date').agg('rating').mean()
```

```python
# How many customers have a membership card?
df.customer_type.value_counts()
```

```python
display(df.groupby('customer_type').count()['invoice_id'].sort_values(ascending=False));

df.groupby('customer_type').count()['invoice_id'].sort_values(ascending=False).plot(kind='bar', color='#66D1C5');
plt.title('Customer Types');
```

```python
# How many female and male customers purchase at the supermarket?
df.gender.value_counts()
```

```python
display(df.groupby('gender').count()['invoice_id'].sort_values(ascending=False));

df.groupby('gender').count()['invoice_id'].sort_values(ascending=False).plot(kind='bar', color='#66D1C5');
plt.title('Customers by Gender');
```

```python
# What supermarket has the highest amount of customers?
df.city.value_counts()
```

```python
display(df.groupby('city').count()['invoice_id'].sort_values(ascending=False));

df.groupby('city').count()['invoice_id'].sort_values(ascending=False).plot(kind='bar', color='#66D1C5');
plt.title('Customers per Supermarket');
```

```python
# What is most frequent payment?
df.type_of_payment.value_counts()
```

```python
display(df.groupby('type_of_payment').count()['invoice_id'].sort_values(ascending=False));

df.groupby('type_of_payment').count()['invoice_id'].sort_values(ascending=False).plot(kind='bar', color='#66D1C5');
plt.title('Type of Payment');
```

```python
# Payment preferences by gender?
sns.histplot(data=df, x='type_of_payment', hue ='gender', palette=['#FF96C2', '#66D1C5']);
```

```python
# Split 'purchase_date' into 'year', 'month' and 'weekday' colums for detailed insights.
```

```python
df["year"] = df.purchase_date.dt.year
```

```python
df["month"] = df.purchase_date.dt.month
```

```python
df["weekday"] = df.purchase_date.dt.weekday
```

```python
df
```

```python
# Rearrange colums
df=df[['invoice_id', 'city', 'product_category', 'unit_price', 'quantity', 'total', 'type_of_payment', 'purchase_date', 'year', 'month', 'weekday', 'time', 'customer_type', 'gender', 'rating']]
```

```python
df
```

```python
df.groupby('month').agg('total').sum()
```

```python
# Total Sales per month: 
# Peak purchase month is January. The dataset only has values from January until March.
display(df.groupby('month').agg('total').sum().sort_values(ascending=False));

df.groupby('month').agg('total').sum().sort_values(ascending=False).plot(kind='bar', color='#66D1C5');
plt.title('Total Sales per Month');
```

```python
# Total Sales per Weekday:
# Peak purchase by weekday (0=monday and 6=sunday) - Peak purchase is on Saturday
display(df.groupby('weekday').agg('total').sum().sort_values(ascending=False));

df.groupby('weekday').agg('total').sum().sort_values(ascending=False).plot(kind='bar', color='#66D1C5');
plt.title('Total Sales per Weekday');
```

```python
# Total Quantity per month:
df.groupby('month').agg('quantity').sum()
```

```python
display(df.groupby('month').agg('quantity').sum().sort_values(ascending=False));

df.groupby('month').agg('quantity').sum().sort_values(ascending=False).plot(kind='bar', color='#66D1C5');
plt.title('Total Quantity per Month');
```

```python
# Purchase behavior by gender
sns.histplot(data=df, x='weekday', hue='gender', palette=['#FF96C2', '#66D1C5'])
```

```python
# Unstack to create a heatmap to display correlations between most bought product category and weekdays.
weekday_and_pc = df.groupby(['product_category','weekday']).agg('total').sum().unstack()
```

```python
sns.heatmap(weekday_and_pc, cmap=sns.color_palette('blend:#FF96C2,#66D1C5', as_cmap=True));
plt.title('Product Category by Weekday');
```
